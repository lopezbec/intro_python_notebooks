#Undertanding the Univariate Linear Regression Model


## Lets create some random data

```python
import numpy as np
np.random.seed(12)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

## Select some values for $\theta_0$ & $\theta_1$

Select some values for $\theta_0$ & $\theta_1$  and run the cell to see the Linear Regression model fitted

```python

Theta_0=4 #@param {type:"integer", min:0, max:14, step:0.5}
Theta_1=6 #@param {type:"integer",step:0.5}
new_y=[t1+Theta_0 for t1 in [x *Theta_1  for x in X]]
plt.plot(X, new_y, color='blue', linewidth=3)
plt.plot(X, y, "b.")
plt.text(0.25, 12, "$h(x)$={0}+ {1}$x$".format(Theta_0,Theta_1), fontsize=16, color="g")
plt.axis([0, 2, 0, 15])
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.show()
    
```

## Cost Function  $J(\theta)$

The objective of linear regression is to minimize the cost function

$$ J(\theta) = \frac{1}{2m} \sum_{i=1}^m \left( h_{\theta}(x^{(i)}) - y^{(i)}\right)^2$$

Select some values for $\theta_0$ & $\theta_1$  and run the cell to try to manually minimize the cost functions

```python
Theta_0=4 #@param {type:"integer", min:0, max:14, step:0.5}
Theta_1=3 #@param {type:"integer",step:0.5}

h_x=[]
for i in range(len(X)):
    h_x_i=Theta_0+(Theta_1*X[i])
    h_x.append(float((h_x_i-y[i])**2))

cost=(1/(2*len(X)))*(sum(h_x))
new_y=[t1+Theta_0 for t1 in [x *Theta_1  for x in X]]
plt.plot(X, new_y, color='blue', linewidth=3)
plt.plot(X, y, "b.")
plt.text(0.25, 13, "$h(x)$={0}+ {1}$x$".format(Theta_0,Theta_1), fontsize=16, color="g")
plt.text(0.25, 11, "J(theta)={}".format(round(cost,3)), fontsize=16, color="g")
plt.axis([0, 2, 0, 15])
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.show()
    
```
